```python
import openpyxl
from openpyxl.styles import PatternFill, colors
```

```python
wb = openpyxl.load_workbook("./日报表_copy.xlsx")
风险控制指标情况_流水 = wb["风险控制指标情况_流水"]
私募种子基金持仓日报表_流水 = wb["私募种子基金持仓日报表_流水"]
风险控制指标情况_标准 =wb["风险控制指标情况_标准"]
私募种子基金持仓日报表_标准 = wb["私募种子基金持仓日报表_标准"]
```

```python
start_row = 1
end_row = 31
start_col = 1
end_col = 8

pd_row = 0
pd_col = 0

for row in range(start_row, end_row+1):
    for col in range(start_col, end_col+1):
        position = (row, col)
        # v1 = 私募种子基金持仓日报表_流水.cell(row, col).value
        # v2 = 私募种子基金持仓日报表_标准.cell(row-pd_row, col-pd_col).value    

        v1 = 风险控制指标情况_流水.cell(row, col).value
        v2 = 风险控制指标情况_标准.cell(row-pd_row, col-pd_col).value  
        if v1 != v2: 
            if type(v1) == float and type(v2) == float:
                if round(v1, 4) == round(v2, 4):  continue
            # 私募种子基金持仓日报表_流水.cell(row, col).fill = PatternFill("solid", fgColor="1874CD")
            # 私募种子基金持仓日报表_标准.cell(row, col).fill = PatternFill("solid", fgColor="FFA300")

            风险控制指标情况_流水.cell(row, col).fill = PatternFill("solid", fgColor="1874CD")
            风险控制指标情况_标准.cell(row-pd_row, col).fill = PatternFill("solid", fgColor="FFA300")
            print(position, "流水表为: ", v1,"标准表为: ", v2)
```
